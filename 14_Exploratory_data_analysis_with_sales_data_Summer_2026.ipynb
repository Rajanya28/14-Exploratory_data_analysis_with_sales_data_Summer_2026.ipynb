### Question 1: Import Data and Store-wise Revenue Analysis (4 Marks)

Import `pandas` and `numpy`. Download the dataset `sales.csv` from https://drive.google.com/drive/folders/1gieHICVDBbUKMZiSF4YRQMAJic-w50JM?usp=sharing. Load the dataset `sales.csv` into a DataFrame. Then, find the top 5 stores with the highest total sales and the bottom 5 stores with the lowest total sales.

**Tasks:**
- Aggregate total sales by store_id
- Rank stores by total revenue
- Compare average sales per department within those stores

```python

import pandas as pd
import numpy as np

# Reading the dataset
sales_data = pd.read_csv("sales.csv")

# Format numbers for better readability
pd.options.display.float_format = '{:,.2f}'.format

# Calculate total revenue generated by each store
revenue_per_store = sales_data.groupby("store_id")["weekly_sales"].agg("sum")

# Sort stores based on revenue
revenue_rank = revenue_per_store.sort_values(ascending=False)

# Extract top and bottom performing stores
highest_revenue_stores = revenue_rank.iloc[:5]
lowest_revenue_stores = revenue_rank.iloc[-5:]

print("===== TOP 5 STORES =====")
print(highest_revenue_stores)

print("\n===== BOTTOM 5 STORES =====")
print(lowest_revenue_stores)

# Department-wise average sales for top stores
top_department_analysis = (
    sales_data.query("store_id in @highest_revenue_stores.index")
              .groupby(["store_id", "department"])
              .weekly_sales.mean()
)

print("\n===== DEPARTMENT-WISE AVERAGE SALES (TOP STORES) =====")
print(top_department_analysis)

# Department-wise average sales for bottom stores
bottom_department_analysis = (
    sales_data.query("store_id in @lowest_revenue_stores.index")
              .groupby(["store_id", "department"])
              .weekly_sales.mean()
)

print("\n===== DEPARTMENT-WISE AVERAGE SALES (BOTTOM STORES) =====")
print(bottom_department_analysis)

print("\nSummary:")
print(f"Highest Revenue Store : {highest_revenue_stores.idxmax()}")
print(f"Lowest Revenue Store  : {lowest_revenue_stores.idxmin()}")
print("Top-ranked stores show greater average departmental sales compared to the lower-ranked stores.")
```

### Question 2: Department Performance Consistency (3 Marks)

Which departments have the most stable weekly sales and which are the most volatile?

**Tasks:**
- Use standard deviation and coefficient of variation
- Identify highly stable and highly volatile departments

```python

import pandas as pd
import numpy as np

# Load dataset
sales_data = pd.read_csv("sales.csv")

# Calculate department statistics
dept_performance = (
    sales_data.groupby("department")["weekly_sales"]
    .agg(["mean", "std"])
)

# Coefficient of Variation
dept_performance["cv"] = (
    dept_performance["std"] / dept_performance["mean"]
) * 100

# Most stable departments (lowest CV)
stable_departments = dept_performance.sort_values("cv").head(5)

# Most volatile departments (highest CV)
volatile_departments = dept_performance.sort_values(
    "cv", ascending=False
).head(5)

pd.options.display.float_format = '{:,.2f}'.format

print("\nMost Stable Departments")
print(stable_departments.to_string())

print("\nMost Volatile Departments")
print(volatile_departments.to_string())

print("\nConclusion:")
print("Departments with lower CV values have more consistent weekly sales.")
print("Departments with higher CV values experience greater fluctuations in sales.")
```

### Question 3: Holiday vs Non-Holiday Sales (3 Marks)

Analyze whether holidays significantly affect weekly sales.

**Tasks:**
- Compare average sales during holidays vs non-holidays
- Calculate percentage increase/decrease
- Identify departments benefiting most from holidays

```python
import pandas as pd
import numpy as np

# Load dataset
sales_data = pd.read_csv("sales.csv")

pd.options.display.float_format = '{:,.2f}'.format

# Average sales during holidays and non-holidays
holiday_avg = sales_data.loc[sales_data["is_holiday"] == True, "weekly_sales"].mean()
non_holiday_avg = sales_data.loc[sales_data["is_holiday"] == False, "weekly_sales"].mean()

# Percentage change
sales_change = ((holiday_avg - non_holiday_avg) / non_holiday_avg) * 100

print("Average Weekly Sales During Holidays :", round(holiday_avg, 2))
print("Average Weekly Sales During Non-Holidays :", round(non_holiday_avg, 2))
print("Percentage Change :", round(sales_change, 2), "%")

# Department-wise holiday impact
department_holiday = (
    sales_data.groupby(["department", "is_holiday"])["weekly_sales"]
    .mean()
    .unstack()
)

department_holiday.columns = ["Non-Holiday Avg", "Holiday Avg"]

department_holiday["Holiday Impact (%)"] = (
    (department_holiday["Holiday Avg"] - department_holiday["Non-Holiday Avg"])
    / department_holiday["Non-Holiday Avg"]
) * 100

# Top departments benefiting from holidays
top_beneficiaries = department_holiday.sort_values(
    by="Holiday Impact (%)",
    ascending=False
).head(5)

print("\nDepartments Benefiting Most From Holidays")
print(top_beneficiaries)

print("\nConclusion:")
if sales_change > 0:
    print(f"Overall sales increase by {sales_change:.2f}% during holidays.")
else:
    print(f"Overall sales decrease by {abs(sales_change):.2f}% during holidays.")

print("The departments listed above experience the greatest positive impact from holiday periods.")
```

### Question 4: Seasonal Trend Detection (3 Marks)

Identify whether sales increase or decrease during specific months.

**Tasks:**
- Extract month from the date column
- Compute average monthly sales
- Determine highest and lowest sales months

```python
import pandas as pd
import numpy as np

# Read dataset
sales_data = pd.read_csv("sales.csv")

# Convert date column to datetime format
sales_data["date"] = pd.to_datetime(sales_data["date"])

# Extract month name
sales_data["month"] = sales_data["date"].dt.month_name()

# Calculate average monthly sales
monthly_sales = (
    sales_data.groupby("month")["weekly_sales"]
    .mean()
)

# Arrange months in calendar order
month_order = [
    "January", "February", "March", "April",
    "May", "June", "July", "August",
    "September", "October", "November", "December"
]

monthly_sales = monthly_sales.reindex(month_order)

pd.options.display.float_format = '{:,.2f}'.format

print("Average Monthly Sales")
print(monthly_sales)

# Highest and lowest sales months
best_month = monthly_sales.idxmax()
best_month_sales = monthly_sales.max()

worst_month = monthly_sales.idxmin()
worst_month_sales = monthly_sales.min()

print("\nHighest Sales Month")
print(f"{best_month}: {best_month_sales:,.2f}")

print("\nLowest Sales Month")
print(f"{worst_month}: {worst_month_sales:,.2f}")

print("\nConclusion:")
print(f"Sales peak during {best_month} with an average weekly sales value of {best_month_sales:,.2f}.")
print(f"The lowest sales occur during {worst_month} with an average weekly sales value of {worst_month_sales:,.2f}.")
print("This indicates a seasonal pattern in customer purchasing behavior.")
```

### Question 5: Store and Department Contribution (4 Marks)

Which combinations of store and department contribute the most to overall revenue?

**Tasks:**
- Group using store_id and department
- Find top and worst-performing combinations
- Compute percentage contribution

```python
import pandas as pd
import numpy as np

# Load dataset
sales_data = pd.read_csv("sales.csv")

pd.options.display.float_format = '{:,.2f}'.format

# Calculate revenue for each Store-Department combination
store_department_sales = (
    sales_data.groupby(["store_id", "department"])["weekly_sales"]
    .sum()
    .reset_index(name="total_revenue")
)

# Total revenue across all combinations
overall_revenue = store_department_sales["total_revenue"].sum()

# Percentage contribution
store_department_sales["contribution_percent"] = (
    store_department_sales["total_revenue"] / overall_revenue
) * 100

# Top 5 combinations
top_combinations = (
    store_department_sales
    .sort_values(by="total_revenue", ascending=False)
    .head(5)
    .reset_index(drop=True)
)

# Bottom 5 combinations
bottom_combinations = (
    store_department_sales
    .sort_values(by="total_revenue")
    .head(5)
    .reset_index(drop=True)
)

print("===== TOP 5 STORE-DEPARTMENT COMBINATIONS =====")
print(top_combinations.to_string(index=False))

print("\n===== BOTTOM 5 STORE-DEPARTMENT COMBINATIONS =====")
print(bottom_combinations.to_string(index=False))

print("\nConclusion:")
print("• The top Store-Department combinations contribute the largest share of overall revenue.")
print("• Department 20 appears frequently among the highest-performing combinations.")
print("• The lowest-performing combinations contribute only a very small percentage of total revenue.")
print("• These results help identify the strongest and weakest revenue-generating business units.")
```

### Question 7: K-Means Clustering (5 Marks)

Import `KMeans` from `sklearn.cluster`. Initialize and fit the K-Means clustering model on the MNIST features (`X`).
Since we know there are 10 digits (0-9), set the number of clusters to 10. Assign the cluster labels to a variable `kmeans_labels`.

**Expected Output:** A successfully fitted K-Means model and the cluster labels array.

```python
from sklearn.datasets import load_digits
from sklearn.cluster import KMeans

# Load dataset
digits = load_digits()

# Features and labels
X = digits.data
y = digits.target

# Initialize K-Means with 10 clusters
kmeans = KMeans(n_clusters=10, random_state=42)

# Fit model and get cluster labels
kmeans_labels = kmeans.fit_predict(X)

print("K-Means model fitted successfully!")
print("\nCluster Labels:")
print(kmeans_labels)
```

### Question 8: F1 Score Evaluation for Clustering (5 Marks)

Evaluate the clustering performance using the F1 score. Since K-Means assigns arbitrary cluster labels, you will first need to map each cluster label to the most frequent true label in that cluster.
After mapping the labels, calculate and print the macro-averaged F1 score using `sklearn.metrics.f1_score`.

**Expected Output:** The calculated F1 score of the clustering.

```python
from sklearn.datasets import load_digits
from sklearn.cluster import KMeans
from sklearn.metrics import f1_score
from scipy.stats import mode
import numpy as np

# Load dataset
digits = load_digits()
X = digits.data
y = digits.target

# K-Means clustering
kmeans = KMeans(n_clusters=10, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

# Map cluster labels to actual digit labels
label_mapping = {}

for cluster in range(10):
    cluster_points = y[kmeans_labels == cluster]

    if len(cluster_points) > 0:
        label_mapping[cluster] = mode(cluster_points, keepdims=True).mode[0]

# Create mapped predictions
predicted_labels = np.array(
    [label_mapping[label] for label in kmeans_labels]
)

# Calculate Macro F1 Score
f1 = f1_score(y, predicted_labels, average='macro')

print("Macro F1 Score:", round(f1, 4))
```
